1 - Załadowanie biblioteki matplotlib i innych potrzebnych bibliotek

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

2 - załadowanie danych 

```python
df = pd.read_csv("IHME_GBD_2019_SMOKING_TOB_1990_2019_NUM_SMOKERS_Y2021M05D27.csv", encoding = "utf-8") 
df
```

3 - użycie matplotlib

```python
matplot = df[(df["sex_name"] == "Both") & (df["location_name"] == "Poland")]
matplot
```

```python
plt.plot(matplot["year_id"], matplot["val"])
```

6 - wykresy liniowe 

```python
liniowy_series = df[(df["sex_name"] == "Both") & (df["location_name"] == "Poland")]
liniowy_series
```

7 - wykresy kolumnowe

```python
fig, axes = plt.subplots(2, 1)
data = df[(df["sex_name"] == "Both") & (df["location_name"] == "Poland")]
data = pd.Series(liniowy_series["val"])
data.plot.bar(ax=axes[0], color='b', alpha=0.7)
data.plot.barh(ax=axes[1], color='g', alpha=0.7)
```

```python
w_kolumnowy = df[(df["sex_name"] == "Both") & (df["location_name"] == "Poland")]
w_kolumnowy = w_kolumnowy[["upper", "lower"]]
w_kolumnowy = w_kolumnowy.head(10)
w_kolumnowy
```

```python
w_kolumnowy.plot.bar()
```

```python
w_kolumnowy.plot.barh(stacked=True, alpha=0.5)
```

8 - seaborn

```python
import seaborn as sns
```

```python
sb = df
sb = sb[sb["year_id"]> 2000]
sb
```

```python
sns.barplot(x='val', y='year_id', data=sb, orient='h')
```

```python
sns.barplot(x='val', y='year_id', hue= "sex_name", data=sb, orient='h')
```

9 - histogramy i wykresy gęstosci rozkładu

```python
histogram = df[df["year_id"] > 2000]
histogram = histogram["val"]
histogram
```

```python
histogram.plot.hist(bins=50)
```

```python
histogram.plot.kde()
```

10 - wykresy punktowe i wykresy bitowe

```python
punkt = df
punkt = punkt[punkt["year_id"] > 2000]
punkt
```

```python
sns.regplot(x='upper', y='lower', data=punkt)
```

```python
sns.pairplot(punkt, diag_kind='kde', plot_kws={'alpha': 0.2})
```

11 - dane kategoryczne

```python
katego = df
katego = katego[(katego["location_name"] == "Poland")]
katego
```

```python
sns.catplot(x='year_id', y='val', hue='age_group_id', col='sex_name', kind='bar', data=katego)
```

```python
sns.catplot(x='year_id', y='val', row='age_group_id', col='sex_name', kind='bar', data=katego)
```
